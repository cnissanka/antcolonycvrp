```python
from collections import OrderedDict

class Node(object):
    def __init__(self, distance, demand):
        self.distance = distance; 
        self.demand = demand 
        self.nonClustered = True; 
    
    def getDemand(self):
        return self.demand
    
    def getDistance(self):
        return self.distance 
    
    def __str__(self):
        return str(self.demand) + " " + str(self.distance)   
    
def getDistancesFromNodeToAllNodes(node, distanceMatrix, isNonClusterMap): 
    distances = distanceMatrix.iloc[:, node] 
    result = {} 
    for index, distance in enumerate(distances):
        if (distances[index] != 0 and isNonClusterMap[index]):
            result[index] = Node(distance, data[index][1]) 
    
    items = OrderedDict(sorted(result.items(), key=lambda x: x[1].distance, reverse=False))
    
    return items 
```

```python
def markAsNonClustered(data):
    return createNodeMap(data, True) 

def createNodeMap(data, flag):
    nodeMap = {} 
    for index, d in enumerate(data): 
        nodeMap[index] = flag 
    return nodeMap 

def createDemandMap(data):
    demandMap = {} 
    for index, d in enumerate(data):
        demandMap[index] = float(d[1])
    return demandMap 

def areThereNonClusterNodes(clusterMap, size=82):
    index = 0; 
    while (index < size):
        if clusterMap[index]:
            return True
        index += 1 
    return False 
```

```python
def createClusterFor(index, isClusterMap, demandMap, capacity):
    clusterPoints = [] 
    iterationCount = 1
    while (areThereNonClusterNodes(isClusterMap)):       
        distancesFromIndex = getDistancesFromNodeToAllNodes(index, distanceMatrix, isClusterMap) 
        newCluster = [index]
        isClusterMap[index] = False 
        totalDemand = demandMap[index]
        distanceKeys = list(distancesFromIndex.keys()); 
        distanceIndex = 0; 
        while (totalDemand <= capacity and distanceIndex < len(distanceKeys)):
            distanceKey = distanceKeys[distanceIndex]
            totalDemand += demandMap[distanceKey]
            isClusterMap[distanceKey] = False; 
            newCluster.append(distanceKey) 
            distanceIndex += 1; 
        clusterPoints.append(newCluster)
    
        if (distanceIndex >= len(distanceKeys)):
            break;
        else:
            index = distanceIndex;
        print("created cluster {}".format(iterationCount)) 
        iterationCount += 1 
    return clusterPoints

CAPACITY = 3500
numberOfNodes = len(data)
isClusteredMap = markAsNonClustered(data)
demandMap = createDemandMap(data) 
clusters = []; 

for k in range(numberOfNodes):
    cluster = createClusterFor(k, isClusteredMap, demandMap, CAPACITY);
    isClusteredMap = markAsNonClustered(data);  
    clusters.append(cluster)

clusters
```

```python
# find best cluster based on the convex hull of ith cluster / number of nodes in that cluster and take the sum
from scipy.spatial import ConvexHull
def createDataMap(data):
    dataMap = {} 
    for index, d in enumerate(data): 
        longitude, latitude = float(d[2]), float(d[3])
        dataMap[index] = [longitude, latitude] 
    return dataMap

def getPointCoordinates(index,data):
    longitude = float(data[index][2])
    latitude = float(data[index][3])
    return [longitude, latitude] 

def getClusterConvexHullArea(points):
    if (len(points) > 2):
        return ConvexHull(points).area
    return 0

def getPointsFromCluster(dataMap, cluster):
    points = [] 
    if not (0 in cluster):
        cluster.insert(0,0)
    for index in cluster:
        p = dataMap[index] 
        points.append(p) 
    return points 
        

def findParameter(dataMap, clusters):
    param = 0 
    for cluster in clusters:
        area = getClusterConvexHullArea(getPointsFromCluster(dataMap, cluster)) 
        numberOfNodes = len(cluster)
        param += (area/numberOfNodes) 
    return param 

dataMap = createDataMap(data)
top3clusters = [20, 20, 20] 
for i in range(len(clusters)):
    param = findParameter(dataMap, clusters[i]) 

    if (top3clusters[0] > param):
        top3clusters[0] = i 
    elif (top3clusters[1] > param): 
        top3clusters[1] = i 
    elif (top3clusters[2] > param): 
        top3clusters[2] = i
        



```

```python
#ant colony optimizer
import pants
import math
import random

memo = {} 

def euclidean(a, b):
    return math.sqrt(pow(a[1] - b[1], 2) + pow(a[0] - b[0], 2))

def getIndexMap(data):
    p = {} 
    for index, d in enumerate(data): 
        longitude, latitude = float(d[2]), float(d[3])
        p[(longitude, latitude)] = index 
    return p 

def getPoints(dataMap, arr):
    l = [] 
    for i in arr:
        l.append((dataMap[i][0], dataMap[i][1]))  
    return l 


def getOptimalRoute(indexMap, dataMap, cluster, alpha, beta, rho, ants):
    nodes = getPoints(dataMap, cluster) 
    world = pants.World(nodes, euclidean) 
    solver = pants.Solver(alpha=alpha, beta=beta, rho=rho, ant_count=ants) 
    solution = solver.solve(world) 
    tour = [] 
    for coord in solution.tour: 
        tour.append(indexMap[coord])
    return tour 

import time

def findOptimalRoute(data,alpha, beta, rho, ants):
    start = time.time()

    optimalRoutes = []
    indexMap = getIndexMap(data)
    for cluster in clusters[top3clusters[0]]:
        if 0 in cluster:
            cluster.remove(0) 
        optimalRoutes.append(getOptimalRoute(indexMap, dataMap, cluster,alpha, beta, rho, ants))

    end = time.time()
    print(f"Runtime of the program is {end - start}")
    return optimalRoutes

    

antsSize = [10, 20, 30, 40, 50] 
rhos = [0.5,0.6, 0.8, 0.9]
betas = [1,1.5,2,2.5,3] 
alphas = [0.7,0.8,0.9,1] 

```

```python
# test all sub ant colony
optimalSubRoutes = []
for a in antsSize:
    for r in rhos: 
        for b in betas: 
            for al in alphas: 
                start = time.time()
                opt = findOptimalRoute(data, al, b, r, a)
                end = time.time()
                optimalSubRoutes.append([opt, (end-start)]) 
```
