```python
import pickle 
with open('routesSubRes.pkl', 'rb') as f:
    subroutes_distances = pickle.load(f)
with open('routesMainRes.pkl', 'rb') as f: 
    mainroute_distances = pickle.load(f) 
```

```python
stats1 = [] 

for rec in subroutes_distances:
    params,computeTime,distanceTravel = rec[0], rec[1][-1], rec[-1]
    stats1.append([params, computeTime, distanceTravel])
```

```python
stats1
```

```python
stat2 = []
for rec in mainroute_distances: 
    a, b, c = rec[0], rec[2], rec[-1] 
    stat2.append([a,b,c])

stat2
```

```python
import pandas as pd 

df1 = pd.DataFrame(stats1)
df2 = pd.DataFrame(stat2)

df2
```

```python
# distance in cluster based ant colony
df1.sort_values(2, kind='mergesort', inplace=False)
```

```python
# time in cluster based ant colony
df1.sort_values(1, kind='mergesort', inplace=False)
```

```python
# distance in tsp first then cluster based ant colony
df2.sort_values(1, kind='mergesort', inplace=False)
```

```python
# time in tsp first then cluster based ant colony
df2.sort_values(2, kind='mergesort', inplace=False)
```
